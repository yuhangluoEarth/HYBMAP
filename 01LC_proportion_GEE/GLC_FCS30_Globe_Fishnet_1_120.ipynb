```python
def generateGrid(xmin, xmax, dx, y1, y2):
    xx = ee.List.sequence(xmin, ee.Number(xmax).subtract(dx), dx)
    cells = xx.map(func_uke).flatten()
    return ee.FeatureCollection(cells)

def func_uke(x):
    x1 = ee.Number(x)
    x2 = ee.Number(x).add(ee.Number(dx))
    coords = ee.List([x1, y1, x2, y2])
    rect = ee.Algorithms.GeometryConstructors.Rectangle(coords)
    return ee.Feature(rect)
```

```python
def func_feh(feature):
    #自然植被                    
    Rainfedcropland5 = feature.getNumber('Rainfedcropland5')
    Herbaceouscover5 = feature.getNumber('Herbaceouscover5')
    TreeorshrubcoverOrchard5 = feature.getNumber('TreeorshrubcoverOrchard5')
    Irrigatedcropland5 = feature.getNumber('Irrigatedcropland5')
    Openevergreenbroadleavedforest5 = feature.getNumber('Openevergreenbroadleavedforest5')
    Closedevergreenbroadleavedforest5 = feature.getNumber('Closedevergreenbroadleavedforest5')
    Opendeciduousbroadleavedforest5 = feature.getNumber('Opendeciduousbroadleavedforest5')
    Closeddeciduousbroadleavedforest5 = feature.getNumber('Closeddeciduousbroadleavedforest5')
    Openevergreenneedleleavedforest5 = feature.getNumber('Openevergreenneedleleavedforest5')
    Closedevergreenneedleleavedforest5 = feature.getNumber('Closedevergreenneedleleavedforest5')
    Opendeciduousneedleleavedforest5 = feature.getNumber('Opendeciduousneedleleavedforest5')
    Closeddeciduousneedleleavedforest5 = feature.getNumber('Closeddeciduousneedleleavedforest5')
    Openmixedleafforest5 = feature.getNumber('Openmixedleafforest5')
    Closedmixedleafforest5 = feature.getNumber('Closedmixedleafforest5')
    Shrubland5 = feature.getNumber('Shrubland5')
    Evergreenshrubland5 = feature.getNumber('Evergreenshrubland5')
    Deciduousshrubland5 = feature.getNumber('Deciduousshrubland5')
    Grassland5 = feature.getNumber('Grassland5')
    Lichensandmosses5 = feature.getNumber('Lichensandmosses5')
    Sparsevegetation5 = feature.getNumber('Sparsevegetation5')
    Sparseshrubland5 = feature.getNumber('Sparseshrubland5')
    Sparseherbaceous5 = feature.getNumber('Sparseherbaceous5')
    Wetlands5 = feature.getNumber('Wetlands5')
    Impervioussurfaces5 = feature.getNumber('Impervioussurfaces5')
    Bareareas5 = feature.getNumber('Bareareas5')
    Consolidatedbareareas5 = feature.getNumber('Consolidatedbareareas5')
    Unconsolidatedbareareas5 = feature.getNumber('Unconsolidatedbareareas5')
    Waterbody5 = feature.getNumber('Waterbody5')
    Permanenticeandsnow5 = feature.getNumber('Permanenticeandsnow5')
    Filledvalue5 = feature.getNumber('Filledvalue5')
    Land5 = feature.getNumber('Land5')
    
    #feature的面积
    return feature.set({
        #自然植被
        'Rainfedcropland5':Rainfedcropland5.divide(Land5),
        'Herbaceouscover5':Herbaceouscover5.divide(Land5),
        'TreeorshrubcoverOrchard5':TreeorshrubcoverOrchard5.divide(Land5),
        'Irrigatedcropland5':Irrigatedcropland5.divide(Land5),
        'Openevergreenbroadleavedforest5':Openevergreenbroadleavedforest5.divide(Land5),
        'Closedevergreenbroadleavedforest5':Closedevergreenbroadleavedforest5.divide(Land5),
        'Opendeciduousbroadleavedforest5':Opendeciduousbroadleavedforest5.divide(Land5),
        'Closeddeciduousbroadleavedforest5':Closeddeciduousbroadleavedforest5.divide(Land5),
        'Openevergreenneedleleavedforest5':Openevergreenneedleleavedforest5.divide(Land5),
        'Closedevergreenneedleleavedforest5':Closedevergreenneedleleavedforest5.divide(Land5),
        'Opendeciduousneedleleavedforest5':Opendeciduousneedleleavedforest5.divide(Land5),
        'Closeddeciduousneedleleavedforest5':Closeddeciduousneedleleavedforest5.divide(Land5),
        'Openmixedleafforest5':Openmixedleafforest5.divide(Land5),
        'Closedmixedleafforest5':Closedmixedleafforest5.divide(Land5),
        'Shrubland5':Shrubland5.divide(Land5),
        'Evergreenshrubland5':Evergreenshrubland5.divide(Land5),
        'Deciduousshrubland5':Deciduousshrubland5.divide(Land5),
        'Grassland5':Grassland5.divide(Land5),
        'Lichensandmosses5':Lichensandmosses5.divide(Land5),
        'Sparsevegetation5':Sparsevegetation5.divide(Land5),
        'Sparseshrubland5':Sparseshrubland5.divide(Land5),
        'Sparseherbaceous5':Sparseherbaceous5.divide(Land5),
        'Wetlands5':Wetlands5.divide(Land5),
        'Impervioussurfaces5':Impervioussurfaces5.divide(Land5),
        'Bareareas5':Bareareas5.divide(Land5),
        'Consolidatedbareareas5':Consolidatedbareareas5.divide(Land5),
        'Unconsolidatedbareareas5':Unconsolidatedbareareas5.divide(Land5),
        'Waterbody5':Waterbody5.divide(Land5),
        'Permanenticeandsnow5':Permanenticeandsnow5.divide(Land5),
        'Filledvalue5':Filledvalue5.divide(Land5),
        'Land5':Land5.divide(Land5)
     })
```
